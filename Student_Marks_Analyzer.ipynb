```python
import pandas as pd
data = {
    "Name": ["Ali", "Sara", "Bilal", "Ayesha", "Ahmed", "Zara", "Usman", "Hina", "Hamza", "Iqra"],
    "Age": [22, 21, 23, 20, 22, 21, 24, 20, 23, 21],
    "Department": ["CS", "IT", "AI", "ML", "CS", "IT", "AI", "ML", "CS", "AI"],
    "Marks": [88, 92, 75, 85, 90, 78, 82, 89, 67, 94],
    "Attendance": [90, 95, 80, 85, 92, 88, 86, 91, 75, 97]
}
# Create DataFrame
df = pd.DataFrame(data)

# Show DataFrame
print(df)
# Save as CSV dataset
df.to_csv("students.csv", index=False)

print("✅ students.csv dataset file created!")
```

```python
# Load the CSV
df = pd.read_csv("students.csv")  # or your correct path
# Display first few rows
print(df.head())
# Get basic info
print(df.info())
# Get statistics for numerical columns
print(df.describe())
```

```python
# Check missing values
print(df.isnull().sum())
# Check duplicates
print(df.duplicated().sum())
```

```python
# Average marks
print("Average Marks:", df["Marks"].mean())
# Highest and lowest marks
print("Top Student:\n", df.loc[df["Marks"].idxmax()])
print("Lowest Student:\n", df.loc[df["Marks"].idxmin()])
# Average marks by department
print("\nAverage marks by Department:")
print(df.groupby("Department")["Marks"].mean())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
plt.bar(df["Name"], df["Marks"], color="skyblue")
plt.title("Student Marks")
plt.xlabel("Student Name")
plt.ylabel("Marks")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Pie chart — Department distribution

```python
dept_counts = df["Department"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(dept_counts, labels=dept_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Department Distribution")
plt.show()
```

## Scatter plot — Attendance vs Marks

```python
plt.figure(figsize=(7,5))
plt.scatter(df["Attendance"], df["Marks"], color="green")
plt.title("Attendance vs Marks")
plt.xlabel("Attendance (%)")
plt.ylabel("Marks")
plt.grid(True)
plt.show()
```
